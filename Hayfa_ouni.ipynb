## Clustering

```python
# Cluster on 2 features
df2 = df1[['Age','NumberOfBadComments']]
df2.head()
```

```python
#scatter plot
sns.scatterplot(df2['Age'],df2['NumberOfBadComments'])
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
errors = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df2)
    errors.append(kmeans.inertia_)
```

```python
# plot the results for elbow method
plt.figure(figsize=(13,6))
plt.plot(range(1,11),errors)
plt.plot(range(1,11),errors, linewidth=3, color='red', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
km = KMeans(n_clusters=2)
km.fit(df2)
y= km.predict(df2)
df2['label']=y
df2.head()
```

```python
sns.scatterplot(x='Age', y='NumberOfBadComments', data=df2, hue='label', s=50)
```

```python
# Cluster on 3 features
df3 = df1[['Age','NumberOfBadComments','EnvironmentSatisfaction']]
df3.head()
```

```python
# plot the results for elbow method
plt.figure(figsize=(13,6))
plt.plot(range(1,11),errors)
plt.plot(range(1,11),errors, linewidth=3, color='red', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
km = KMeans(n_clusters=2)
km.fit(df3)
y= km.predict(df3)
df3['label']=y
df3.head()
```

```python
#3d scatter plot
fig = plt.figure(figsize=(17,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df3['Age'][df3['label']==0],df3['NumberOfBadComments'][df3['label']==0],df3['EnvironmentSatisfaction'][df3['label']==0], s=50 , c='red')
ax.scatter(df3['Age'][df3['label']==1],df3['NumberOfBadComments'][df3['label']==1],df3['EnvironmentSatisfaction'][df3['label']==1], s=50 , c='green')
ax.scatter(df3['Age'][df3['label']==2],df3['NumberOfBadComments'][df3['label']==2],df3['EnvironmentSatisfaction'][df3['label']==2], s=50 , c='blue')
ax.view_init(40)
ax.set_xlabel('Age')
ax.set_ylabel('No. Bad Comments')
ax.set_zlabel('Environment Satisfaction')
plt.show()
```
